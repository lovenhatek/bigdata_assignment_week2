##**열 삭제하기**

```python
import gdown
gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False)
```

```python
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv', low_memory = False)
ns_df.head()
```

마지막 'Unnamed: 13'열은 CSV파일 각 라인 끝에 콤마가 있어서 판다스가 자동으로 추가한 것이다. 불필요한 열이므로 삭제하면 된다.

```python
ns_book = ns_df.loc[:, '번호':'등록일자']
ns_book.head()
```

중간에 있는 열을 제외하려면 슬라이싱을 사용하기 어렵다. 이때 불리언배열(boolean)을 사용하면 쉽게 원하는 열만 선택할 수 있다.

```python
print(ns_df.columns)
```

columns의 속성은 판다스의 index클래스 객체이다. 이 객체의 원소는 파이썬의 리스트처럼 숫자 인덱스로 참조할 수 있다. 예를 들어 첫 번째 열의 이름인 '번호'를 선택할 때 다음과 같이 쓸 수 있다.

```python
print(ns_df.columns[0])
```

```python
ns_df.columns != 'Unnamed: 13'
```

원소별 비교를 활용하여 'Unnamed: 13'열이 아닌 것을 표시하는 배열을 만들면 위와 같다. 결과는 넘파이(NumPy)배열이다(파이썬 리스트와 비슷한 것으로 생각해도 괜찮음). 원소별 비교 연산으로 불리언배열을 얻었다. 이를 selected_columns변수에 저장하고 판다스 데이터프레임의 loc메서드에 전달하면 True인 열의 행만 선택할 수 있다.

```python
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns] # True인 열의 모든 행을 선택한다
ns_book.head()
```

```python
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

데이터프레임의 메서드를 사용하여 열을 삭제할 수 있다. 판다스에서는 데이터프레임의 행이나 열을 삭제하는 drop메서드를 제공한다. drop메서드로 열을 삭제하려면 첫 번째 매개변수에 삭제하려는 열 이름을 전달하고 axis매개변수를 1로 지정한다.

```python
ns_book = ns_df.drop('Unnamed: 13', axis = 1)
ns_book.head()
```

```python
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis = 1)
ns_book.head()
```

지금까지는 ns_df데이터프레임에서 열을 삭제한 결과를 ns_book이라는 변수에 저장하였다. drop메서드에 inplace매개변수를 True로 지정하면 현재 선택한 데이터프레임을 바로 수정할 수도 있다. 앞서 생성한 ns_book데이터프레임의 '주제분류번호'를 삭제해보겠다.

```python
ns_book.drop('주제분류번호', axis = 1, inplace = True)
ns_book.head()
```

판다스는 비어 있는 값을 NaN으로 표시한다. drop메서드와 비슷한 dropna메서드는 기본적으로 NaN이 하나 이상 포함된 행이나 열을 삭제, axis = 1로 지정하여 ns_df데이터프레임에서 NaN이 포함된 열을 삭제한다. (삭제된 열에는 하나 이상의 NaN이 포함되어 있었다는 이야기)

```python
ns_book = ns_df.dropna(axis = 1)
ns_book.head()
```

모든 값이 NaN인 열을 삭제하려면 dropna메서드에 how 매개변수를 'all'로 지정하면 된다.

```python
ns_book = ns_df.dropna(axis = 1, how = 'all')
ns_book.head()
```

## **행 삭제하기**

행을 삭제할 때도 drop메서드를 사용할 수 있다. axis 매개변수를 0으로 지정하면 행을 삭제할 수 있지만, 기본값이 0이기 때문에 생략해도 괜찮다. 처음 2개 행을 삭제해 보면 다음과 같은 결과가 나온다.

```python
ns_book2 = ns_book.drop([0,1])
ns_book2.head()
```

행을 선택할 때 자주 사용하는 것은 [] 연산자이다. [] 연산자에 열 이름 또는 열 이름 리스트를 전달하여 데이터프레임의 열을 선택할 수 있다.

```python
ns_book2 = ns_book[2:]
ns_book2.head()
```

```python
ns_book2 = ns_book[0:2]
ns_book2.head()
```

슬라이싱 이외에도 불리언배열로 행을 선택할 수 있다. 행을 선택할 때 가장 즐겨 사용하는 방법이다. 비교 연산자를 활용해 원하는 행은 True로 표시하고 제외할 행은 False로 표시한 불리언 배열을 만들어 사용한다.

예를 들어 출판사가 '한빛미디어'인 행만 선택하려면, 비교 연산자를 사용하여 원본 데이터프레임 ns_df에서 출판사가 '한빛미디어'인 행만 True인 불리언 배열을 만든다. 그 다음에 이 불리언 배열을 [] 연산자에 넣어 준다.

```python
selected_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_book[selected_rows]
ns_book2.head()
```

loc메서드에 불리언배열을 사용하여 행을 선택할 수 있다.

```python
ns_book2 = ns_book.loc[selected_rows]
ns_book2.head()
```

```python
ns_book2 = ns_book[ns_book['대출건수'] > 1000]
ns_book2.head()
```

열 삭제할 때와 마찬가지로 행 삭제할 때도 loc메서드, drop메서드, dropna메서드 모두 사용할 수 있지만, 불리언배열을 연산자에 전달하는 방법을 자주 사용한다.

## **중복된 행 찾기**

판다스 데이터프레임의 중복된 행은 duplicated메서드를 사용하여 검사한다. 중복된 행 중에서 처음 행을 제외한 나머지는 True로, 그 외에 중복되지 않은 나머지 모든 행은 False로 표시한 불리언 배열을 반환한다.

sum함수를 함께 사용하면 True를 1로 인식하기 때문에 duplicated메서드로 만든 불리언 배열에서 중복된 행의 개수를 셀 수 있다.

```python
sum(ns_book.duplicated())
```

```python
sum(ns_book.duplicated(subset = ['도서명', '저자', 'ISBN']))
```

ns_book데이터프레임의 모든 행은 '번호' 열에 고유한 값을 가지고 있기 때문에 중복된 행이 나올 수가 없다. 만약 '도서명', '저자', 'ISBN'등 일부 열을 기준으로 중복된 행이 있는지 찾으려면 duplicate메서드의 subset매개변수에 기준열을 나열하면 된다.

'도서명', '저자', 'ISBN'을 기준으로 찾으면 중복된 행이 2만 개가 넘는다. 이들을 기준으로 어떤 데이터가 중복되었는지 확인해보겠다. duplicated메서드에 keep 매개변수를 False로 지정하여 중복된 모든 행을 True로 표시하게 한다. 이 데이터를 ns_book3 변수에 저장하여 확인해보면 다음과 같은 결과가 나온다.

```python
dup_rows = ns_book.duplicated(subset = ['도서명', '저자', 'ISBN'], keep = False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

## **그룹별로 모으기**





어떤 도서가 인기가 많을지 예상하려고 하므로 이 데이터프레임에서는 '대출건수'열이 중요하다. 따라서 같은 도서의 대출건수를 하나로 합치면 편하다.

예를 들어, '혼공데이터분석' 도서가 두 권이라 하나로 합치려 한다면, groupby메서드를 사용하면 된다.

groupby메서드의 by매개변수는 행을 합칠 때 기준이 되는 열을 지정한다.
권수가 한 개 이상인 도서는 각기 다른 도서로 간주한다.
예를 들어 파친코 1권과 2권을 별개의 도서로 본다.

'도서명','출판사','발행연도','대출건수'열을 기준으로 행을 합치면 다음과 같다.

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
print(count_df)
```

결과를 변환한 count_df 데이터프레임에 groupby메서드를 적용하고, 정수 타입인 '대출건수' 열은 sum메서드를 사용한다.
NaN이 포함되어 있는 행을 삭제하고 계산하면 대출건수 합계에서 빠지기 때문에 이를 막기 위해 dropna 매개변수를 False로 지정한다.

```python
group_df = count_df.groupby(by = ['도서명', '저자', 'ISBN', '권'], dropna = False)
loan_count = group_df.sum()
print(loan_count)
```

```python
loan_count = count_df.groupby(by = ['도서명', '저자', 'ISBN', '권'],
                              dropna = False).sum()
loan_count.head()
```
